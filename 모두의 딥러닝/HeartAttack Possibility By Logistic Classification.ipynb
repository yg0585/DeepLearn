```python
import tensorflow as tf
import numpy as np

xy = np.loadtxt("/content/heart.csv", delimiter=',', dtype= np.float32, encoding='utf-8-sig')
x_data = xy[1:, 0:-1]
y_data = xy[1:, [-1]]

print(x_data.shape, y_data.shape)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = x_data.shape[1], activation ='sigmoid'))
tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.005),  metrics=['accuracy'])
tf.model.summary()
```

```python
history = tf.model.fit(x_data, y_data, epochs = 1000)
```
